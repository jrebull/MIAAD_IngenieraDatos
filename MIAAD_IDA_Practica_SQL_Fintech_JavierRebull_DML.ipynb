## <font color="#003CA6">Introducción</font>

El lenguaje SQL (Structured Query Language) es el estándar universal para la gestión y manipulación de datos en bases de datos relacionales. En el contexto de la industria fintech, donde las operaciones transaccionales requieren precisión, trazabilidad y velocidad, el dominio de las sentencias DML (Data Manipulation Language) resulta fundamental para cualquier ingeniero de datos.

Las operaciones DML comprenden cuatro instrucciones principales: `SELECT` para consultar datos, `INSERT` para agregar registros, `UPDATE` para modificar valores existentes y `DELETE` para eliminar filas de forma controlada. Según Reis y Housley (2022), el ingeniero de datos debe garantizar que cada transformación aplicada sobre los datos mantenga la integridad referencial y responda a los objetivos de negocio, especialmente en entornos financieros donde un error en una consulta puede tener impacto directo en reportes regulatorios o en la experiencia del cliente.

Esta práctica aplica dichas operaciones sobre una base de datos SQLite que simula el ecosistema transaccional de una fintech con cuentas en MXN y USD, abarcando desde la exploración básica de tablas hasta la modificación y eliminación segura de registros.

## Ejercicio 1. Exploración básica


- Lista todas las filas de la tabla **clientes**.  
- Muestra únicamente el nombre y la ciudad de cada cliente.  
- ¿Cuántos clientes hay en total?  


````python
# 1a) Todas las filas de clientes:
%%sql
SELECT * FROM clientes;
````

````python
# 1b) Solo nombre y ciudad:
%%sql
SELECT nombre, ciudad FROM clientes;
````

````python
# 1c) Total de clientes:
%%sql
SELECT COUNT(*) AS total_clientes FROM clientes;
````

## Ejercicio 2. Cuentas de clientes


- Obtén todas las cuentas del cliente **C001**.  
- Lista las cuentas que están en moneda **USD**.  
- Muestra el número de cuentas que tiene cada cliente (usa `GROUP BY`).  


````python
# 2a) Cuentas del cliente C001:
%%sql
SELECT * FROM cuentas WHERE cliente_id = 'C001';
````

````python
# 2b) Cuentas en USD:
%%sql
SELECT * FROM cuentas WHERE moneda = 'USD';
````

````python
# 2c) Número de cuentas por cliente:
%%sql
SELECT cliente_id, COUNT(*) AS num_cuentas
FROM cuentas
GROUP BY cliente_id;
````

## Ejercicio 3. Transacciones


- Obtén todas las transacciones de la cuenta **A1001**.  
- Calcula el **saldo neto** de esa cuenta sumando la columna `monto`.  
- Lista las transacciones que sean **comisiones**.  


````python
# 3a) Transacciones de la cuenta A1001:
%%sql
SELECT * FROM transacciones WHERE cuenta_id = 'A1001';
````

````python
# 3b) Saldo neto de A1001:
%%sql
SELECT SUM(monto) AS saldo_neto FROM transacciones WHERE cuenta_id = 'A1001';
````

````python
# 3c) Transacciones que sean comisiones:
%%sql
SELECT * FROM transacciones WHERE concepto LIKE '%Comisión%';
````

## Ejercicio 4. Reportes con agrupaciones


- Calcula el total de transacciones (suma de `monto`) por **categoría**.     


````python
# 4a) Total de monto por categoría:
%%sql
SELECT categoria, SUM(monto) AS total_monto
FROM transacciones
GROUP BY categoria;
````


## Ejercicio 5. **Modificar datos (UPDATE)**

> **Objetivo:** actualizar valores existentes usando `WHERE` de forma segura.

1. Cambia el `segmento` del cliente **C002** de *Retail* a **PYME**.  
2. Corrige la `ciudad` de **Textiles J&R (C003)** a **Saltillo**.  


> **Tip:** siempre valida primero con un `SELECT` y después aplica el `UPDATE`.


````python
# 5a) Cambiar segmento de C002 a PYME:
%%sql
-- Verificar antes
SELECT * FROM clientes WHERE cliente_id = 'C002';
````

````python
%%sql
-- Aplicar cambio
UPDATE clientes SET segmento = 'PYME' WHERE cliente_id = 'C002';

-- Verificar después
SELECT * FROM clientes WHERE cliente_id = 'C002';
````

````python
# 5b) Corregir ciudad de C003 a Saltillo:
%%sql
-- Verificar antes
SELECT * FROM clientes WHERE cliente_id = 'C003';
````

````python
%%sql
-- Aplicar cambio
UPDATE clientes SET ciudad = 'Saltillo' WHERE cliente_id = 'C003';

-- Verificar después
SELECT * FROM clientes WHERE cliente_id = 'C003';
````


## Ejercicio 6. **Borrar datos (DELETE)**

> **Objetivo:** eliminar filas específicas manteniendo la integridad y usando `WHERE`.

1. Elimina **solo** las transacciones de la cuenta **A3001** del día **2025-07-12**.  
2. Elimina la **comisión** de `A2001` del **2025-07-04**.  
> **Tip:** prueba primero con `SELECT` para verificar el subconjunto a borrar.


````python
# 6a) Eliminar transacciones de A3001 del 2025-07-12:
%%sql
-- Verificar qué vamos a borrar
SELECT * FROM transacciones
WHERE cuenta_id = 'A3001' AND fecha = '2025-07-12';
````

````python
%%sql
-- Borrar
DELETE FROM transacciones
WHERE cuenta_id = 'A3001' AND fecha = '2025-07-12';

-- Confirmar que ya no existen
SELECT * FROM transacciones
WHERE cuenta_id = 'A3001' AND fecha = '2025-07-12';
````

````python
# 6b) Eliminar comisión de A2001 del 2025-07-04:
%%sql
-- Verificar qué vamos a borrar
SELECT * FROM transacciones
WHERE cuenta_id = 'A2001' AND fecha = '2025-07-04' AND concepto LIKE '%Comisión%';
````

````python
%%sql
-- Borrar
DELETE FROM transacciones
WHERE cuenta_id = 'A2001' AND fecha = '2025-07-04' AND concepto LIKE '%Comisión%';

-- Confirmar que ya no existe
SELECT * FROM transacciones
WHERE cuenta_id = 'A2001' AND fecha = '2025-07-04' AND concepto LIKE '%Comisión%';
````

<h3 style="color:#003CA6;">Conclusiones Técnicas</h3>

<p>Esta práctica permitió aplicar de forma integral las operaciones DML sobre una base de datos financiera, reforzando competencias esenciales para el ingeniero de datos en entornos fintech. Los principales aprendizajes son:</p>

<ol>
  <li><b>La exploración previa es la base de toda consulta confiable:</b> Antes de escribir cualquier sentencia, inspeccionar la estructura de las tablas (tipos de dato, claves, cardinalidad) evita errores silenciosos y garantiza que los resultados sean interpretables en contexto de negocio.</li>

  <li><b>GROUP BY transforma datos transaccionales en inteligencia operativa:</b> Agrupar transacciones por categoría o cuentas por cliente convierte registros individuales en métricas accionables, exactamente lo que un área de reporteo regulatorio o comercial necesita para tomar decisiones.</li>

  <li><b>UPDATE y DELETE exigen disciplina quirúrgica:</b> El patrón SELECT antes, modificar después y SELECT para confirmar no es opcional en producción. Una cláusula WHERE mal construida en un entorno bancario real puede afectar miles de registros y generar inconsistencias con impacto regulatorio.</li>

  <li><b>SQL sigue siendo el lenguaje universal de los datos:</b> Independientemente de si se trabaja con SQLite, PostgreSQL o plataformas cloud, la lógica relacional y las sentencias DML son transferibles. El dominio de estos fundamentos es lo que distingue a un ingeniero de datos que entiende el negocio de uno que solo ejecuta scripts.</li>
</ol>
